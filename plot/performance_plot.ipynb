```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt


log_performance_path = "results_single_test/batch_20240314T201633/log_performance_1.csv"
individual_metrics_path = "results_single_test/batch_20240314T201633/df_individual_metrics_1_20240314T201633.csv"

log_performance = pd.read_csv(log_performance_path)
individual_metrics = pd.read_csv(individual_metrics_path)
```

```python
ptsBack = 20
ptsForward = 5

inference = log_performance[(log_performance["datetime"] > str(dt.datetime.fromtimestamp(individual_metrics["start_time"][0] - ptsBack))) & (log_performance["datetime"] < str(dt.datetime.fromtimestamp(individual_metrics["end_time"][0] + ptsForward)))]

inference.plot(x="datetime", y="memUsage")
# points marked indicating model load, start and end
points = pd.DataFrame({"x":[33,41,74], "y":[inference.iloc[33]["memUsage"], inference.iloc[41]["memUsage"], inference.iloc[74]["memUsage"]]})
plt.scatter(x=points["x"], y=points["y"])
plt.show()

inference.plot(x="datetime", y="cpuUsage", c="red")
# points marked indicating model load, start and end
points = pd.DataFrame({"x":[33,41,74], "y":[inference.iloc[33]["cpuUsage"], inference.iloc[41]["cpuUsage"], inference.iloc[74]["cpuUsage"]]})
plt.scatter(x=points["x"], y=points["y"], c="red")
plt.show()
```
